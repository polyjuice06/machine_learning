```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
from itertools import chain
url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
```

```python
html = urlopen(url)
soup = BeautifulSoup(html, features='lxml')
```

```python
headers = [th.getText() for th in soup.find_all('tr')[0].find_all('th')]
```

```python
print(headers)
```

```python
col_name = ['Player','Pos','3P','2P','TRB','AST','STL','BLK']
```

```python
rows = soup.find_all('tr',attrs={"class": "full_table"})
stats = ['player','pos','fg3_per_g','fg2_per_g','trb_per_g',
         'ast_per_g','stl_per_g','blk_per_g']
useful = []
for i in range(len(rows)):
    rows_data = [[td.getText() for td in rows[i].find_all(['td','th'],attrs={'data-stat': stat})] for stat in stats]
    useful.append(list(chain(*rows_data)))
```

```python
useful
```

```python
df = pd.DataFrame(useful, columns = col_name)
df.index = df.index + 1
```

```python
df
```

```python
front_court_df = df[df['Pos'].str.contains('PF|C')].head(50)
front_court_df
```

```python
guard_df = df[df['Pos'].str.contains('SG|PG')].head(50)
guard_df
```

```python
sf_df = df[df['Pos'] == "SF"].head(50)
sf_df
```

```python
using_df = pd.concat([front_court_df,guard_df,sf_df]).reset_index(drop=True)
using_df
```

```python
using_df.to_csv('2020-21_NBA_player_stat.csv')
```
