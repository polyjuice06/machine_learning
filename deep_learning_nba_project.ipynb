```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
from itertools import chain
url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
```

```python
html = urlopen(url)
soup = BeautifulSoup(html, features='lxml')
```

```python
headers = [th.getText() for th in soup.find_all('tr')[0].find_all('th')]
```

```python
print(headers)
```

```python
col_name = ['Player','Pos','TRB','AST']
```

```python
rows = soup.find_all('tr',attrs={"class": "full_table"})
stats = ['player','pos','trb_per_g','ast_per_g']
useful = []
for i in range(len(rows)):
    rows_data = [[td.getText() for td in rows[i].find_all(['td','th'],attrs={'data-stat': stat})] for stat in stats]
    useful.append(list(chain(*rows_data)))
```

```python
useful
```

```python
df = pd.DataFrame(useful, columns = col_name)
df.index = df.index + 1
```

```python
df
```

```python
front_court_df = df[df['Pos'].str.contains('C')].head(10)
front_court_df
```

```python
guard_df = df[df['Pos'].str.contains('PG')].head(10)
guard_df
```

```python
using_df = pd.concat([front_court_df,guard_df]).reset_index(drop=True)
using_df
```

```python
import random
import math
a  = random.uniform(1,-1)
b  = random.uniform(1,-1)
c  = random.uniform(1,-1)
d  = random.uniform(1,-1)
e  = random.uniform(1,-1)
f  = random.uniform(1,-1)
g  = random.uniform(1,-1)
h  = random.uniform(1,-1)
i  = random.uniform(1,-1)

eta = 0.0001

def sigmoid(x):
    return 1.0/(1.0+math.exp(-x))

data = [(using_df.loc[i]['TRB'],using_df.loc[i]['AST']) for i in range(len(using_df))]
labels = [-1 if using_df.loc[i]['Pos'] == 'C' else 1 for i in range(len(using_df))]
zValue = [0 for _ in range(20)]
FinalValue = [0 for _ in range(20)]

for counter in range(100000):
    da=db=dc=dd=de=df=dg=dh=di=0.0
    for idx in range(len(data)):
        x = float(data[idx][0])
        y = float(data[idx][1])
        z1 = a*x+b*y+c
        X = sigmoid(z1)
        z2 = d*x+e*y+f
        Y = sigmoid(z2)
        z = g*X+h*Y+i
        Z = labels[idx]
        dz = 2*(z-Z)
        dz1 = dz*g*sigmoid(z1)*(1.0-sigmoid(z1))
        dz2 = dz*h*sigmoid(z2)*(1.0-sigmoid(z2))
        da+= dz1*x
        db+= dz1*y
        dc+= dz1
        dd+= dz2*x
        de+= dz2*y
        df+= dz2
        dg+= dz*X
        dh+= dz*Y
        di+= dz
        a= a-eta*da
        b = b-eta*db
        c = c-eta*dc
        d = d-eta*dd
        e = e-eta*de
        f = f-eta*df
        g = g-eta*dg
        h = h-eta*dh
        i = i-eta*di
print("a=%lf,b=%lf,c=%lf,d=%lf,e=%lf,f=%lf,g=%lf,h=%lf,i=%lf"%(a,b,c,d,e,f,g,h,i))

for counter in range(100000):
    for idx in range(len(data)):
        x = float(data[idx][0])
        y = float(data[idx][1])
        z1 = a*x+b*y+c
        X=sigmoid(z1)
        z2=d*x+e*y+f
        Y=sigmoid(z2)
        z = g*X+h*Y+i
        zValue[idx]+=z
for idx in range(len(data)):
    FinalValue[idx] = zValue[idx]/100000
print(FinalValue)
```
